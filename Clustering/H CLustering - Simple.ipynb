```
from scipy.spatial import distance_matrix
import numpy as np
import pandas as pd
np.set_printoptions(precision=2)
x = np.array([[0.4,0.53],
              [0.22,0.38],
              [0.35,0.32],
              [0.26,0.19],
              [0.08,0.41],
              [0.45,0.3]])
z= np.tril(np.ones([6,6]),-1)
op = distance_matrix(x,x) *z
df = pd.DataFrame(op,index=['P1','P2','P3','P4','P5','P6'],columns=['P1','P2','P3','P4','P5','P6'])
df
```
